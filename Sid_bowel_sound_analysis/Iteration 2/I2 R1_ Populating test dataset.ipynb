```python
# creating randomized data for columns from test data, into new dataframe using pandas

dj = pd.DataFrame(pd.Series(random.choices(['0','1'], weights=[1, 1], k=100)), columns=['Label(0=Pre/1=Post)'])

#dj['Label(0=Pre/1=Post)'] = pd.Series(random.choices(['0','1'], weights=[1, 1], k=len(dj)))
dj['Age in Days'] = pd.DataFrame(np.random.randint(2, 6, size=(len(dj),1)))
dj['Gender(0=M,1=F)'] = pd.Series(random.choices(['0','1'], weights=[1, 1], k=len(dj)))
dj['Sick Baby(0=N,1=Y)'] = pd.Series(random.choices(['0','1'], weights=[1, 1], k=len(dj)))

dj
```

Creating the start time and end time values

```python
# creating base columns into new dataframe for start and end time
ds = pd.DataFrame((np.random.randint(1,2,size=(100,10))), columns=['S1','E1','S2','E2','S3','E3','S4','E4','S5','E5'])

count = 0

for i in range(1,6):
  ds[f'S{i}'] += count
  count += 1
  ds[f'E{i}'] += count
  count += 1

# creating an array to add subtract random decimals between -0.5 and 0.5 from dataframe to make data look unrelated
plus_minus_array = np.round_(np.random.uniform(-0.5,0.5,size=(100,10)), decimals=2)
ds += plus_minus_array
ds
```

```python
# creating a randomized decimal array between 0.8 and 10 to be multiplied with each row.
# this is done so that there is a random looking difference between the different times
abc = np.round_(np.random.uniform(0.8,10,size=(100)), decimals=2)

# multiplying array with each row of dataframe
ds = ds.mul(abc, axis=0).round(2)
ds
```

```python
djj = ds
```

```python
def recur_it(dsn,value_count):
  """Recursive function to remove random number of bowel sounds from each row"""
  random_int = random.randint(0,99)
  if dsn.loc[random_int, 'E5'] == -987:
    return recur_it(dsn, value_count)
  else:
    if value_count < 80:
      if value_count < 20:
        for i in range(2,10):
          dsn.iat[random_int, i] = -987
      elif value_count < 40:
        for i in range(4,10):
          dsn.iat[random_int, i] = -987
      elif value_count < 60:
        for i in range(6,10):
          dsn.iat[random_int, i] = -987
      elif value_count < 80:
        for i in range(8,10):
          dsn.iat[random_int, i] = -987
      value_count += 1
      return recur_it(dsn, value_count)
    else:
      dsn = dsn.replace(-987, pd.NA)
      return dsn

# running the recursive function on dataframe
djj = recur_it(djj, 0)

# start and end time are now creating with random number of bowel sounds
djj
```

```python
# joining main dataframe, with start and end time dataframe
dfz = dj.join(djj)
dfz
```
