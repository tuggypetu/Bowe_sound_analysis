## Graph plots

```python
# R2: import visualization tools matplotlib and seaborn
import matplotlib.pyplot as plt 
import seaborn as sns

plt.scatter(df['bowelSounds/min'], df['Avg duration'])
plt.xlabel("Bowel Sounds per min") #x label
plt.ylabel("Average duration of a Bowel Sound") #y label
plt.show()
```

Count plots for different attributes-

```python
sns.countplot(x='Age in Days',hue='Gender(0=M,1=F)',data=df)
```

```python
sns.countplot(x='Sick Baby(0=N,1=Y)',data=df)
```

Boxplots to visualize the variance of clinical measurements based on sickness-

```python
plt.figure(figsize=(10, 5))
sns.boxplot(x='Sick Baby(0=N,1=Y)',y='Avg duration',data=df,palette='coolwarm')
```

```python
plt.figure(figsize=(10, 7))
sns.boxplot(x='Sick Baby(0=N,1=Y)',y='bowelSounds/min',data=df,palette='YlGnBu')
```

```python
df_drop = df.drop(['S1','E1','S2','E2','S3','E3','S4','E4','S5','E5'], axis=1)

# heatmap to check if any coorelation between all attributes through visualisation
plt.figure(figsize = (10,5))
sns.heatmap(df_drop.corr(), cmap="Blues", annot = True)
plt.show()
```

'Sick Baby(0=N,1=Y)' and 'bowelSounds/min' seem to have higher correlation with each other from the heatmap.
